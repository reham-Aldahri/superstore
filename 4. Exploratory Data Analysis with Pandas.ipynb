````python
# B. Find the farthest south destination (the result with the lowest latitude)

farthest_south_destination = destinations[0]
farthest_south_lat = geocoder.arcgis(destinations[0]).lat

for destination in destinations:
    current_lat =  geocoder.arcgis(destination).lat
    if current_lat < farthest_south_lat:
        farthest_south_lat = current_lat
        farthest_south_destination = destination

print (farthest_south_destination)
````

````python
# C. For each destination, describe the direction being traveled from the previous destination
#    Hint: You can keep track of where we are using current_lat and current_lng variables
#    If the next destination has a greater latitude, it means we're going North
#    If the next destination has a greater longitude, it means we're going east
#    Print a message in the following format: "Next, we travel SouthWest to XXXXX"

# عرفنا اول متغير عندنا بانه ف= الخط الطولي و العرضي الفلاني
# والي عرفناه با ال latlng

current_location = geocoder.arcgis(destinations[0]).latlng

#  بدأنا اللوب بالفور .. و حددنا المكان ضمن الامكنة الموجودة عندنا 
# وعرفنا المسج الي نطبعه ف الاخير 
#و حددنا اللوكيشين الجديد الي ح يكون المكان الي في عبارة الفور لوب

for destination in destinations:
    message = "Next, we travel "
    new_location = geocoder.arcgis(destination).latlng
    
# نبدأ الشرط عندنا  الي هوا في حال كان خط الطول و الي عبرنا عنده بالاندكس 0  اكبر من المكان الحالي 
# الي حددناه في عبارة الفور لوب 
#وحددنا الطباعة انه الاتجاه ح يكون شمال  والا ح يكون جنوب في حال ما تحقق الشرط

    if(new_location[0] > current_location[0]):
        message += "North"
    else:
        message += "South"

# الشرط الي هنا نفس الي فوق  بس انه استخدمنا خط العرض
#  رايح يمين شرق و رايح يسار غرب 
    if(new_location[1] > current_location[1]):
        message += "East"
    else:
            message += "West"
            
# طيب هنا خرجنا عن عبارة الاف و الان ح نطبع الاوتبوت الي نبغاه من الفور لوب 
# المسج الي كتبناه ف البداية ح ينضاف له كلام جديد بعد تحقق الشروط السابقة 

    message += f" to {destination}"
    
# عند امر الطباعة : اول شي ح تنكتب المسج الي هيا كانت الاتجاهات
# و المنطقة المراد الوصول لها كانت متعرفة في المسج المكتوب  و قعدنا نضيف عليه كلام بتحقق الشروط
# بعدين ح يكتب لنا المكان الحالي الي طبعا ح يتغير مع كل لووب و يسير يكتب لنا اسم المكان الي اخده من الليست
# الي اصلا متعرفة سابقا 
    print(message)
    current_location = new_location
````

```python
import pandas as pd
df= pd.read_csv('orders.csv')
```

```python
# A. What are the three most profitable orders purchased by customer PO-8865?
df[df['customer_id']=='PO-8865' ].sort_values(by = 'profit',ascending=False).head(3)
```

```python
# B. What are the five orders of product_id TEC-HEW-10002304 with the highest quanitity?
df[ df['product_id']=='TEC-HEW-10002304' ].sort_values(by = 'quantity',ascending=False).head(5)
```

```python
pd.DataFrame(df.query('product_id == "TEC-HEW-10002304"').sort_values(by = 'quantity' , ascending = False).head(5))
```
